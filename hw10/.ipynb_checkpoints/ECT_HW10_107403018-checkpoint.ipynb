```python
from sklearn import svm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction import DictVectorizer
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
X=df_all1.drop(['tag','name'],axis=1)
n_features = X.shape[1]
y=df_all1[['tag']].values.ravel()
n_classes = y.shape[0]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
param_grid=[{'C': [1000.0, 5000.0, 10000.0, 50000.0, 100000.0],
             'gamma': [0.0001, 0.0005, 0.001, 0.005, 0.01, 0.1]}]
clf = GridSearchCV(SVC(), param_grid, cv=5, scoring='%s_macro' % 'precision')
clf.fit(X_train, y_train)
means = clf.cv_results_['mean_test_score']

print("Best parameters set found on development set:")
print()
#  列印最佳參數
print(clf.best_params_)
print()
print("Grid scores on development set:")
print()
print("Higest Accuracy:")
print()
print(max(means))
print()
print("Grid scores on development set:")
#  這邊依設置的參數列印出所有參數的得分狀況
for mean, params in zip(means, clf.cv_results_['params']):
    print("%0.7f for %r" % (mean, params))
```

```python
svc=SVC(C=1000,gamma=0.1)
svc.fit(X_train, y_train)

predicted = svc.predict(X_test)

# get the accuracy
accuracy_score(y_test, predicted)
```

```python
# 測試超參數優化
RF_param = {"n_estimators":[10,50,100,150,200],"max_depth":[5,10,15,20]}

# 實例化隨機森林分類器
RF = RandomForestClassifier()
GSCV = GridSearchCV(RF,param_grid=RF_param,cv=5,scoring='%s_macro' % 'precision')
GSCV.fit(X_train,y_train)
means = GSCV.cv_results_['mean_test_score']
print("Best parameters set found on development set:")
print()
#  列印最佳參數
print(GSCV.best_params_)
print()
print("Higest Accuracy:")
print()
print(max(means))
print()
print("Grid scores on development set:")

 #  這邊依設置的參數列印出所有參數的得分狀況
for mean, params in zip(means, GSCV.cv_results_['params']):
    print("%0.7f for %r" % (mean, params))
```

```python
rfc=RandomForestClassifier(n_estimators=50,max_depth=15)
rfc.fit(X_train, y_train)

predicted = rfc.predict(X_test)

# get the accuracy
accuracy_score(y_test, predicted)
```

```python
trainX=X_train.values.reshape(9639,181,1,1)
testX=X_test.values.reshape(3214,181,1,1)
```

```python
import keras
from keras import layers
from keras.models import Sequential  #用來啟動 NN
from keras.layers import Conv2D  # Convolution Operation
from keras.layers import MaxPooling2D # Pooling
from keras.layers import Flatten
from keras.layers import Dense # Fully Connected Networks
from keras.layers import Dropout

model = Sequential()  
model.add(Conv2D(16,(1,1), input_shape = (181,1,1), activation = 'relu'))
model.add(MaxPooling2D(pool_size =(1,1)))
#Second convolutional layer
model.add(Conv2D(36, (1,1), activation = 'relu'))
model.add(MaxPooling2D(pool_size = (1, 1)))
model.add(Flatten()) 
model.add(Dense(units = 128, activation = 'relu'))
model.add(Dropout(0.25))
model.add(Dense(units = 10, activation = 'relu'))
model.add(Dense(units = 1, activation = 'softmax'))
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
model.summary()
```

```python
train_history=model.fit(x=trainX,y=y_train,validation_split=0.2, epochs=10, batch_size=300,verbose=2)
```

```python
score = model.evaluate(testX,y_test)
print("Test loss:", score[0])
print("Test accuracy:", score[1])
```
