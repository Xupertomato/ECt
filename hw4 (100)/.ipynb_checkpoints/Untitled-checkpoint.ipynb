```python
import pandas as pd # 引用套件並縮寫為 pd  
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
pd.set_option('display.width',1000)
df = pd.read_csv('/Users/xwlee/Desktop/ECt/hw5/ageinc.csv') 
```

```python
df.insert(2, column="z_income", value=preprocessing.scale(df['income']))
df.insert(3, column="z_age", value=preprocessing.scale(df['age']))
df
```

```python
df1 = df.drop(['income','age'],axis=1)
df1
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2,init="k-means++",n_init=10,max_iter=300,tol=1e-04,random_state=10)
kmeans.fit(df1)
print(kmeans.inertia_)
```

```python
sse_list = [ ] 
K = range(2, 10) 
for k in range(2,10): 
    kmeans=KMeans(n_clusters=k) 
    kmeans.fit(df1) 
    sse_list.append(kmeans.inertia_)   #model.inertia_返回模型的误差平方和，保存进入列表
    print(sse_list)

#折线图展示聚类数——SSE曲线
plt.figure() 
plt.plot(np.array(K), sse_list, 'bx-')
plt.rcParams['figure.figsize'] = [12,8]
plt.xlabel('k',fontsize=18)
plt.ylabel('sum of Squares',fontsize=18)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
